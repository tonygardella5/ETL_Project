## Extract

```python
# Import dependencies
import pandas as pd
```

```python
# Read csv file into pandas
file = "../Resources/Space_Corrected.csv"
df = pd.read_csv(file)
df
```

## Transform

```python
# Remove extra columns
df = df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"])
df.head()
```

```python
# Data count
df['Company Name'].value_counts()
```

```python
# Create DataFrame for SpaceX
SpaceX_df = df.loc[df['Company Name']== "SpaceX"]
SpaceX_df.head()
```

```python
# Save to csv
SpaceX_df.to_csv("../Resources/SpaceX_info.csv")
```
