## Extract

```python
# Import dependencies
import pandas as pd
import datetime as dt
import numpy as np
```

```python
# Read json file into pandas
file = "../Resources/News_Category_Dataset_v2.json"
df = pd.read_json(file, lines=True)
df
```

## Transform

```python
# substring to be searched 
sub ='SpaceX'
  
# creating and passsing series to new column 
df["Indexes"]= df["headline"].str.find(sub) 
  
# display 
df
```

```python
df['Indexes'].value_counts()
```

```python
df = df[df.Indexes != -1]
df
```

```python
# Save df to csv file
df.to_csv('../Resources/News_Reports.csv', index=False)
```

```python
# Sort by date range
bins = [np.datetime64(date(2012,1,1)),np.datetime64(date(2013,1,1)),np.datetime64(date(2014,1,1)), np.datetime64(date(2015,1,1)), np.datetime64(date(2016,1,1)),np.datetime64(date(2017,1,1)),np.datetime64(date(2018,1,1)),np.datetime64(date(2019,1,1)),np.datetime64(date(2020,1,1))]
group_names = ['2012','2013','2014', '2015', '2016', '2017', '2018', '2019']
#df['date'] = pd.to_datetime(df['date'])
df['date'] = pd.to_datetime(df['date'])
df['categories'] = pd.cut(df['date'], bins, labels=group_names)
print(df)
```
