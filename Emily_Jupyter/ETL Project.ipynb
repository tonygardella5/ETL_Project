## ETL Project

```python
# Import dependencies
import pandas as pd
import datetime as dt
import requests
import json
```

## Source :  Spaceflight News
Website:  https://spaceflightnewsapi.net

```python
# URL for GET requests to retrieve vehicle data
url = "https://spaceflightnewsapi.net/api/v1/articles"
```

```python
# Pretty print JSON for all articles
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# Narrow search to SpaceX
url_2 = "https://spaceflightnewsapi.net/api/v1/articles?search=spacex"

# Pretty print JSON for all articles
response = requests.get(url_2).json()
print(json.dumps(response, indent=4, sort_keys=True))    
```

```python
# Store responses
json_responses = []

#Loop through all responses
for x in range (int(100)):
    r = requests.get(url_2)
    json_responses.append(r.json()) 
    
```

```python
# Create dataframe to hold json responses
df = pd.DataFrame(json_responses)
df
```
