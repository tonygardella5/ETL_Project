## Extract

```python
# Import dependencies
import pandas as pd
import datetime as dt
import requests
import json
import datetime as dt
import numpy as np
```

```python
# URL for GET requests to retrieve SpaceX news
json_responses = []

for page in range(359):
    url = f'https://spaceflightnewsapi.net/api/v1/articles?page={page}/articles?spacex'
    response = requests.get(url).json()
    r = requests.get(url).json()
    json_responses.append(r)
```

```python
# Store responses
json_responses = []

#Loop through all responses
for x in range (20):
    r = requests.get(url).json()
    json_responses.append(r) 
    
    
# Store responses
#json_responses = []

#Loop through all responses
#for x in range (25):
#    r = response.get(url)
#    json_responses.append(r.json()) 
   
    
#type(json_responses)    
```

```python
# Lists for specific data
title = []
published_date = []
featured_image = []
launches = []
news_sites = []
url = []


# Loop through all the responses and append lists
for j in json_responses:
    for i in json_responses:
        title.append(json_responses[0]['docs'][i]['title'])
        published_date.append(json_responses[0]['docs'][i]["published_date"])
        featured_image.append(json_responses[0]['docs'][i]["featured_image"])
        launches.append(json_responses[0]['docs'][i]["launches"])
        news_sites.append(json_responses[0]['docs'][i]["news_site"])
        url.append(json_responses[0]['docs'][i]["url"])
   
```

```python
json_responses
```

```python
# Create DataFrame out of appended lists
df = pd.DataFrame({"Title": title,
                 "Published Date": published_date,
                  "Featured Image": featured_image,
                  "Launches": launches,
                  "News Sites": news_sites,
                  "URL": url})
df
```

```python
# Check data types
df.dtypes
```

```python
# Convert Published Date type from object to datetime
df["Published Date"] = pd.to_datetime(df["Published Date"])
```

```python
# Check Published Date data types
df["Published Date"].dtypes
```

```python
# Sort by date in ascending order
df = df.sort_values(["Published Date"])
df
```
