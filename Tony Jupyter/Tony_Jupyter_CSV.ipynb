```python
bins = [np.datetime64(date(2012,1,1)),np.datetime64(date(2013,1,1)),np.datetime64(date(2014,1,1)), np.datetime64(date(2015,1,1)), np.datetime64(date(2016,1,1)),np.datetime64(date(2017,1,1)),np.datetime64(date(2018,1,1)),np.datetime64(date(2019,1,1)),np.datetime64(date(2020,1,1))]
group_names = ['2012','2013','2014', '2015', '2016', '2017', '2018', '2019']

#df['date'] = pd.to_datetime(df['Datetime'])
df['categories'] = pd.cut(df['Datetime'], bins, labels=group_names)
print(df)
```
